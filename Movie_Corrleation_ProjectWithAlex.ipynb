# Cleaning The Data

```python
#let see if we are missing
for col in df.columns:
        pct_missing = np.mean(df[col].isnull())
        print('() -{}%'.format(col,pct_missing))
```

```python
#looks like we do not have any missing data
#next we look at data types for our columns
df.dtypes
```

```python
#change data type of columns
##df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
```

```python
#there are two columns for year and year released. let's make the year correct
df['yearcorrect']=df['released'].astype(str).str[:4]
df.head()
```

```python
#oreder by gross revenue
df=df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# lets get all of the data first
pd.set_option('display.max_rows', None)
# and then we run the previous line of code again. 
#So doing that we can see all the data
```

```python
#Remove Duplicates
#shows the disticnt values
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# lets try some hypothesis.
# Budget might have high corelation with gross revenue
# company have high corelation with gross revenue
```

```python
# we can use scatter plot to investigate that:

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for film')
plt.show()
```

```python
df.head()
```

```python
#cane we determine if the data is correlated.
# we gonna use regression plot using Seaborn
# Plot the budget vs growth using Seaborn
sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},
            line_kws={"color":"blue"})
```

```python
#now let's determine what the real correlation is
df.corr(method='pearson')  #pearson, kendall,spearman 
#let's go with default
```

```python
#high correlation between Budget and Gross
```

```python
corelation_matrix= df.corr(method='pearson')
sns.heatmap(corelation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Next let's look at company
#Company is not Numeric BUT we can make it as Numeric.
# We can use Numeric Identifier so we can use it in Corrleation Matrix

df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]=df_numerized[col_name].astype('category')
        df_numerized[col_name]=df_numerized[col_name].cat.codes
df_numerized.head()
```

```python
# Doing so, now every string in original table is represented 
#by number (so is numerized)
# So now let's use the corelation matrix again
```

```python
corelation_matrix= df_numerized.corr(method='pearson')
sns.heatmap(corelation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Now every item has a numeric representation and 
# consequently the matrix is much larger
# For a little bit better visulaizatuon
df_numerized.corr()
```

```python

correlation_mat =df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr=sorted_pairs[(sorted_pairs)>0.5]
high_corr
```
